# Agrupamento com C-means (Fuzzy):
* Agrupamento Suave
* Agrupamento em forma de associação não absoluta, de maneira percentual

```python
from sklearn import datasets
import numpy as np
from sklearn.metrics import confusion_matrix
import skfuzzy
```

```python
## Dataset
iris = datasets.load_iris()
```

```python
## Cluster
cluster = skfuzzy.cmeans(data=iris.data.T, c=3, m=2, error=0.005,
                        maxiter=1000, init=None)
```

```python
cluster[1]
```

```python
predict_percents = cluster[1]
```

## Probabilidades

```python
for i in range(150):
    print(predict_percents[0][i], predict_percents[1][i], predict_percents[2][i] )
```

```python
# Matriz de contigência para comparação com as classes originais da base de dados
predict_ = predict_percents.argmax(axis=0)
result = confusion_matrix(iris.target, predict_)
result
```
