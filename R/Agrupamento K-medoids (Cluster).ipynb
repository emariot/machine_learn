# Agrupamento com K-medoids (Clusters):
* Semelhante ao k-means, mas os centros onde se inicia a formação dos clusters são pontos de dados, não pontos aleatórios como acontece no k-means
* Mais robusto

```r
library(cluster)
library(factoextra)
```

```r
cluster = pam(iris[1:4],k=3)
```

```r
cluster
```

```r
plot(cluster)
```

```r
table(iris$Species, cluster$clustering)
```

```r
g = fviz_cluster(list(data=iris[1:4], cluster=cluster$cluster), ellipse.type="norm", ggtheme=theme_bw(), main='Cluster')
plot(g)
```
