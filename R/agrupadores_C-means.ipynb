# Agrupamento com C-means (Fuzzy):
* Agrupamento Suave
* Agrupamento em forma de associação não absoluta, de maneira percentual

```r
library(factoextra)
library(e1071)
library(RColorBrewer)
```

## Cluster Fuzzy

```r
cluster = cmeans(iris[,1:4], centers=3)
cluster$membership
```

```r
heatmap(cluster$membership, Colv=NA, Rowv=NA, col=brewer.pal(n=9, name='YlOrRd'))
```

## Matriz de confusão

```r
table(iris$Species, cluster$cluster)
```

```r
g = fviz_cluster(list(data=iris[1:4], cluster=cluster$cluster), ellipse.type="norm",, ggtheme=theme_bw())
plot(g)
```
