# Agrupadores: Cluter Hierárquico
---
* O metodo hierárquico de análise de cluster tem como principal característica um algorítmo capaz de fornecer mais de um tipo de partição dos dados.
* Gera vários agrupamentos possíveis, onde um cluster pode ser mesclado a outro em determinado passo do algoritmo.

```r
library(factoextra)
```

# Criando Cluster

```r
hc = hcut(iris[,-5], k=3, hc_method="complete", stand=T)
hc
```

## Dendograma

```r
plot(hc)
```

## Dendograma

```r
g1 = fviz_dend(hc, show_labels = FALSE, rect=TRUE, ggtheme=theme_bw(), main='Cluster Hierérquico')
plot(g1)
```

```r
g2 = fviz_cluster(hc, ellipse.type='norm', ggtheme=theme_bw(),
                 main='CLuster Hierárquico')
plot(g2)
```

```r
plot(g1)
plot(g2)
```
