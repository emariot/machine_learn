## Padronização

```r
iris_par = scale(iris[,1:4])
iris_par
```

## Normalização

```r
norm = function(x){
    return((x-min(x))/(max(x)-min(x)))
}
```

```r
iris_norm = norm(iris[,1:4])
iris_norm
```

## Comparando modelos

```r
boxplot(iris[,1:4], main="Dados Originais")
boxplot(iris_par[,1:4], main="Padronizados")
boxplot(iris_norm[,1:4], main="Normalizados")
```
